```python

#파이썬 환율
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

df = pd.DataFrame(columns=['country_name',
                           'buying_cost',
                           'bfee_ratio','selling_cost',
                           'slfee_ratio',
                           'sending_cost',
                           'sefee_ratio',
                           'dealing_br'])

pd.DataFrame(columns=['Region','Rate'])
html = urlopen("https://www.mibank.me/exchange/bank/index.php?search_code=005") # 하나은행
soup = BeautifulSoup(html, 'html.parser') # 홈페이지 크롤링

country_name        = soup.select('tbody > tr > td.first > a')
buying_cost         = soup.select('tbody > tr > td:nth-child(3)')
bfee_ratio          = soup.select('tbody > tr > td:nth-child(4)')
selling_cost        = soup.select('tbody > tr > td:nth-child(5)')
slfee_ratio         = soup.select('tbody > tr > td:nth-child(6)')
sending_cost        = soup.select('tbody > tr > td:nth-child(7)')
sefee_ratio         = soup.select('tbody > tr > td:nth-child(8)')
dealing_br          = soup.select('tbody > tr > td:nth-child(9)')

for i in range(0, len(country_name)):
    df2 = pd.DataFrame({'country_name':[country_name[i].text],
                        'buying_cost':[buying_cost[i].text],
                        'bfee_ratio':[bfee_ratio[i].text.replace("%","",1)],
                        'selling_cost':[selling_cost[i].text],
                        'slfee_ratio':[slfee_ratio[i].text[:-1]],
                        'sending_cost':[sending_cost[i].text],
                        'sefee_ratio':[sefee_ratio[i].text[:-1]],
                        'dealing_br':[dealing_br[i].text]
                      })
    
    df = df.append(df2)
    df.to_excel(excel_writer='sample.xlsx') 

df
```
